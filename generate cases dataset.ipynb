```python
list2=[]
for categories in cat_df['Sub Category']:
    if categories == '-':
        continue;
        
    list2.append(categories)
    list1 = categories.split()
    string1 = "+".join(map(str,list1))
#     print(string1)
    url = r'https://indiankanoon.org/search/?formInput='+string1
#     print(url)
    f = requests.get(url)
    soup = BeautifulSoup(f.content,'lxml')
    list_temp=[]
    for link in soup.findAll('a', href=True, text='Full Document'):
        name = link['href'].split('/')[2]
        print(name)
        list_temp.append(name)
        url_temp = r'https://indiankanoon.org'+link['href']
        payload={'type': 'pdf'}
        files=[

        ]
        headers = {}
        response = requests.request("POST", url_temp, headers=headers, data=payload, files=files)
        with open('webdata/'+name+'.pdf','wb') as f:
            f.write(response.content)
    list2.append(list_temp)
    
pd.DataFrame(list2).to_csv('metadata_webdata.csv', index=False, header=False)
```
